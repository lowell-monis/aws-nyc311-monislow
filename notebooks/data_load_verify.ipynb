```python
import pandas as pd
import boto3
from io import StringIO

s3 = boto3.client('s3')
bucket_name = 'cmse492-monislow-nyc311-975049958391-us-east-1-an'
file_name = 'modeling/data_resolution_time.csv'

obj = s3.get_object(Bucket=bucket_name, Key=file_name)
data = obj['Body'].read().decode('utf-8')
df = pd.read_csv(StringIO(data))
```

```python
# Inspection of Modeling Data
# First few rows
print("First few rows:")
print(df.head())

# Data types and missing values
print("\n" + "="*50)
print("Data types and missing values:")
print(df.info())
```

```python
# Summary statistics
print("Summary statistics:")
print(df.describe())
```

```python
# Check target balance
print("Target distribution (resolved_quickly):")
print(df['resolved_quickly'].value_counts().sort_index())

# As percentage
print("\nTarget distribution (%)")
print((df['resolved_quickly'].value_counts() / len(df) * 100).sort_index().round(2))
```
